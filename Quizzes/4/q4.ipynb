```python
x = np.arange(0,10*np.pi,0.1)   # start,stop,step
y = np.sin(x)
#z=np.sin(x)
z=np.cos(x)
```

```python
plt.plot(x,z, label="cos(x)")
plt.plot(x,y, label="sin(x)")
plt.legend()
plt.show()
```

```python
x = np.arange(0,10*np.pi,0.1)   # start,stop,step
y = -np.sin(x)
z=np.sin(x)
#z=np.cos(x)
```

```python
plt.plot(x,z, label="sin(x)")
plt.plot(x,y, label="-sin(x)")
plt.legend()
plt.show()
```

Now use an FFT package in the language of your choice to compute the cross-correlation between the two resulting arrays from sin() and cos(). Plot the resulting cross-correlation to obtain an illustration like the following.

```python
#when y,z found using -sin(x),sin(x)
result = np.correlate(y, z, mode='full')
lags = np.arange(-x[-1],x[-1]+0.1,0.1) #adding 0.1 to include the last instant of time also
plt.figure()
plt.plot(lags,result, color='black')
plt.xlabel('Lag')
plt.ylabel('Cross-Corr(CCF)')
plt.show()
```

Now compute the autocorrelation of each of the arrays separately and plot the resulting autocorrelations to compare with the CCF between the two as computed in the above. Plot the resulting autocorrelations to obtain illustrations like the following,



```python
import matplotlib.pyplot as plt
import numpy as np

result = np.correlate(y, y, mode='full')
lags = np.arange(-x[-1],x[-1]+0.1,0.1) #adding 0.1 to include the last instant of time also
plt.figure()
plt.plot(lags,result , color='r')
plt.xlabel('Lag')
plt.ylabel('Auto-Corr(CCF): sin(x)')
plt.show()
```

```python
result = np.correlate(z, z, mode='full')
lags = np.arange(-x[-1],x[-1]+0.1,0.1) #adding 0.1 to include the last instant of time also
plt.figure()
plt.plot(lags,result)
plt.xlabel('Lag')
plt.ylabel('Auto-Corr(CCF): cos(x)')
plt.show()
```
