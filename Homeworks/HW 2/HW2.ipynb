```python
#downloads file from given url
def download(url, fpath):
    import requests
    r = requests.get(url)
    with open(fpath, 'wb') as f:
        f.write(r.content)
    return None

#fetching ids txt file
def fetchTriggersfromWeb(filePath="./triggers.txt",
                         url= 'https://raw.githubusercontent.com/cdslaborg/DataRepos_SwiftBat/master/triggers.txt'):
    download(url,filePath)
    return filePath

def readFile(fpath):
    import pandas as pd
    return pd.read_csv(fpath,header=None, dtype='str') 

#creating list of path urls
def path_url(lst, url):
    paths=[]
    for trigger in lst:
        paths.append(url+'GRB'+trigger+'_ep_flu.txt')
    return paths
```

```python
import pandas as pd
import requests
import os
import urllib

def fetchDatafromWeb(triggers,file_urls):
    """
        Downloads .txt file from a corresponding url path in list of urls to folder "files".
        Download skipped if URL file is already on folder, file is empty, or url raises error.
    
    """
    counter=0
    missing=0
    empty=0
    
    for i,url in enumerate(file_urls):
        counter+=1
        fpath='./files/GRB'+triggers[i]+'_ep_flu.txt'
        try:
            if os.path.exists(fpath):
                print(fpath+ "file already in directory. Skipping download...\n")
            else:
                df=pd.read_csv(url)
                if df.size!=0:
                    print("Dowloading file for ID:"+triggers[i])
                    download(url,fpath)
                else:
                    empty+=1
                    print('{} File is empty. Skipping download...'.format(fpath))
                    
        except urllib.request.HTTPError:
            missing+=1
            print('\n url:{} does not exist.'.format(url))
            print('file for GRB ID:{} is missing. {} files missing of {} events so far\n'.format(triggers[i],missing+empty,counter))
            pass
        
    print('\nDownloading completed. Total downloads: {}, Missing files: {}, Total files: {} \n'.format(counter-missing-empty,missing+empty,counter))
```

```python
fetchDatafromWeb(triggers,file_urls)
```

```python
from scipy.stats import kurtosis, skew 
```

```python
class Struct(): pass
def data_summary(inpath="files"):
    
    stat=Struct()
    stat.ndata=0; #counter
    stat.logpeak=Struct() 
    stat.logfluence=Struct()
    
    #moments
    stat.logpeak.mean= []
    stat.logpeak.std=[]
    stat.logpeak.skew=[]
    stat.logpeak.kurt=[]
    
    stat.logfluence.mean =[]
    stat.logfluence.std=[]
    stat.logfluence.skew=[]
    stat.logfluence.kurt=[]
    
    for file in os.listdir(inpath)[1:]:
                                   
        data = np.loadtxt(os.path.join(inpath,file), skiprows=1)   
        if data.size!=0 and all(data[:,1]<0.0): #checks file not empty nor sec column values bigger than 0
            stat.ndata += 1
            data[:,0]= np.log(data[:,0]) #transform first-column nat values to log values
            
            stat.logpeak.mean.append(np.mean(data[:,0]))
            stat.logpeak.std.append(np.std(data[:,0]))
            stat.logpeak.skew.append(skew(data[:,0]))
            stat.logpeak.kurt.append(kurtosis(data[:,0]))
            
            stat.logfluence.mean.append(np.mean(data[:,1]))
            stat.logfluence.std.append(np.std(data[:,1]))
            stat.logfluence.skew.append(skew(data[:,1]))
            stat.logfluence.kurt.append(kurtosis(data[:,1]))
    return stat
    
```

```python
def plot_errorbar(stat,figfile='SwiftData_ReducedErrorPlot.png'):
    
    ax = plt.figure(figsize=1.5*np.array([6.4,4.8]), dpi =150)
    ax = plt.subplot()

    ax.set_xlabel(r"Log Fluence [ ergs/cm$^2$ ]")
    ax.set_ylabel(r" Log Epeak [ keV ]")
    ax.axis (np.log([1.0e-8, 1.0e-1, 1.0, 1.0e4])) #xmin, xmax, ymin, ymax
    
    
    ax.errorbar (
                    stat.logfluence.mean,
                    stat.logpeak.mean,
                    yerr=stat.logpeak.std,
                    xerr=stat.logfluence.std,
                    elinewidth = 1,
                    alpha = 0.4,
                    ecolor = 'r',
                    fmt = 'o'
                )
 
    ax.set_title('Plot of Epeak vs. Fluence for {} Swift GRB events'.format(stat.ndata))
    plt.savefig(figfile)
    plt.show()
    
```

```python
# calculates the covariance between X ---Epeak (Ep) and y---LEiso from given data
def cov(x, y):
 
    mean_x, mean_y = x.mean(), y.mean()
    n = len(x)
    
    return sum((x - mean_x) * (y - mean_y)) / n
 
    
# calculates the Covariance matrix for input data 
def cov_matrix(data):
 
    # get the rows and cols
    rows, cols = data.shape
 
    # the covariance matrix has a shape of n_features x n_features
    # n_features  = cols - 1 (not including the target column)
    cov_mat = np.zeros((cols, cols))
 
    for i in range(cols):
 
        for j in range(cols):
            # store the value in the matrix
            cov_mat[i][j] = cov(data[:, i], data[:, j])
 
    return cov_mat
 
# calculates Correlation matrix using Covariance function (cov()) for input data
def corr_matrix(data):

    rows, cols = data.shape

    corr_mat = np.zeros((cols, cols))

    for i in range(cols):

        for j in range(cols):

            x, y = data[:, i], data[:, j]
            # note: covariance matrix is just being normalized
            corr_mat[i][j] = cov(x, y) / (x.std() * y.std())

    return corr_mat
```

```python
class Struct(): pass
def get_MeanCorrCov(inpath="files"):
    
    stat=Struct()
    stat.ndata=0; #counter
    
    stat.corr= []
    stat.cov =[]
    stat.mean=[]
    
    
    
    for file in os.listdir(inpath)[1:]:
                                   
        data = np.loadtxt(os.path.join(inpath,file), skiprows=1)   
        if data.size!=0 and all(data[:,1]<0.0): #checks file not empty nor sec column values bigger than 0
            stat.ndata += 1
            data[:,0]= np.log(data[:,0]) #transform first-column nat values to log values
            
            stat.mean.append(np.mean(data, axis=0))
            stat.cov.append(cov_matrix(data))
            stat.corr.append(corr_matrix(data))
                    
    return stat

dstat=get_MeanCorrCov()
```

```python
def getEllipsoidBoundary( covMat
                        , meanVec
                        , npoint = 50
                        ):
    """
    
    Return the coordinates of the boundary of an 
    ellipsoid represented by the input ``covMat``
    covariance matrix and the ``meanVec`` center.

        **Parameters**

            covMat

                The representative covariance matrix 
                of the ellipsoid.

            meanVec

                A vector representing the center 
                of the ellipsoid.

            npoint

                The number of points with which the 
                ellipsoid is represented. The more,
                the higher the resolution will be.
                The default value is 50.

        **Returns**

            A matrix of size ``(2, npoint)`` that contains 
            the boundary points of the generated ellipsoid 
            corresponding to the input matrix. 

    """

    from scipy import linalg as la

    independentVariable = np.linspace(0, 2*np.pi, npoint, endpoint = True)
    ap = np.zeros( (2, npoint) )
    ap[0,:] = np.cos(independentVariable)
    ap[1,:] = np.sin(independentVariable)
    eigenValues, eigenVecMat = la.eig(covMat)
    eigenValMat = np.eye(2)
    eigenValMat[0,0] = np.sqrt(eigenValues[0].real)
    eigenValMat[1,1] = np.sqrt(eigenValues[1].real)
    meanVecRepMat = np.ones( (2,npoint) )
    meanVecRepMat[0,:] = meanVec[0]
    meanVecRepMat[1,:] = meanVec[1]
    return np.matmul( np.matmul(eigenVecMat.real , eigenValMat) , ap ) + meanVecRepMat 
```

```python
import seaborn as sns
sns.set_theme(style="white")

def plot_2DEllipses(stat, figfile='SwiftDataPlotEllipse.png'):
    
    #Visualizing data summary
    ax = plt.figure(figsize=1.5*np.array([6.4,4.8]), dpi =150)
    ax = plt.subplot()
    ax.set_xlabel(r"Log Fluence [ ergs/cm$^2$ ]")
    ax.set_ylabel(r" Log Epeak [ keV ]")
    ax.axis (([1.0e-8, 1.0e-1, 1.0, 1.0e4])) #xmin, xmax, ymin, ymax
    ax.set_xscale("log")
    ax.set_yscale("log")

    #### set up color mapping via correlations
    cmap = Struct()
    cmap.values = []
    for i in range(0,stat.ndata): cmap.values.append(stat.corr[i][0,1])
    cmap.values = np.array(cmap.values)
    cmap.name = "coolwarm"

    import matplotlib.colors as colors
    import matplotlib.cm as cmx

    cmapObject = plt.get_cmap(cmap.name)
    #cNorm = colors.Normalize( vmin = np.min(cmap.values), vmax = np.max(cmap.values), clip = True )
    cNorm = colors.Normalize( vmin=-1., vmax=1., clip = True )
    mappable = cmx.ScalarMappable( norm = cNorm, cmap = cmapObject )
    cmap._rgba = np.zeros( ( len(cmap.values) , 4) )
    for i, rho in enumerate(cmap.values): cmap._rgba[i,:] = mappable.to_rgba(cmap.values[i])

    
    for i in range(0,stat.ndata):
        bcrd= getEllipsoidBoundary(stat.cov[i],stat.mean[i])
        ax.plot (
                    np.exp(bcrd[1,:]), #e^log(n)--->n: values for lEipso
                    np.exp(bcrd[0,:]), #e^log(n)--->n: values for Ep
                    color =cmap._rgba[i],
                    alpha=0.25,
                    linewidth= 0.7
                )
    ax.set_title('Plot of Epeak vs. Fluence for {} Swift GRB events'.format(stat.ndata))
    plt.savefig(figfile)
    plt.show()   
```
